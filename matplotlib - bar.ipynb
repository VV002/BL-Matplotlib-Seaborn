1. Write a Python programming to display a bar chart of the popularity of programming

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
#plt.figure(figsize=(10, 6))
plt.bar(languages, popularity, color='blue', label='Popularity of programming')
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.legend()

plt.show()
```

9.  Write a Python programming to display a bar chart of the popularity of programming

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.figure(figsize=(10, 6))  
plt.bar(languages, popularity, color='green')

plt.subplots_adjust(bottom=0.2)  

plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')

plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.

```python
groups = ['Group A', 'Group B', 'Group C']
men_scores = [85, 90, 78]
women_scores = [88, 85, 82]

x = np.arange(len(groups))
bar_width = 0.35

plt.bar(x - bar_width / 2, men_scores, width=bar_width, label='Men', color='skyblue')
plt.bar(x + bar_width / 2, women_scores, width=bar_width, label='Women', color='pink')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(x, groups)
plt.legend()

plt.show()
```

11. Write a Python program to create bar plot from a DataFrame

```python
data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 2],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3]
}
df = pd.DataFrame(data)

df.plot(kind='bar')

plt.title('Bar Plot from DataFrame')
plt.xlabel('Index')
plt.ylabel('Values')

plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure

```python
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4']

plt.bar(labels, mean_velocity, yerr=std_velocity, capsize=5, color='skyblue', edgecolor='black')

plt.title('Mean Velocity with Error Bars')
plt.xlabel('Samples')
plt.ylabel('Mean Velocity')

plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).

```python
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4']

plt.bar(labels, mean_velocity, yerr=std_velocity, capsize=5, color='skyblue', edgecolor='black')

for i, v in enumerate(mean_velocity):
    plt.text(i, v + 0.02, str(int(round(v * 1000))), ha='center', va='bottom')

plt.title('Mean Velocity with Error Bars')
plt.xlabel('Samples')
plt.ylabel('Mean Velocity')

plt.show()
```
